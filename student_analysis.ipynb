```python
df = pd.read_csv(r"C:\Users\Umang Tiwari\OneDrive\Documents\student-mat.csv", sep=",")
print("Data load ho gaya!")
print("Rows aur Columns:", df.shape)
df.head()
```

```python
print("Missing values:")
print(df.isnull().sum())
print("\nDuplicates:", df.duplicated().sum())
df = df.drop_duplicates()
print("Data clean ho gaya!")
print("Data types:")
print(df.dtypes)
```

```python
# Average final grade
print("Average Final Grade (G3):", df['G3'].mean())

# 15 se zyada score karne wale
print("15 se upar students:", len(df[df['G3'] > 15]))

# Gender performance
print("\nGender average grades:")
print(df.groupby('sex')['G3'].mean())
```

```python
# Graph 1 - Grades ka histogram
plt.figure(figsize=(8,5))
plt.hist(df['G3'], bins=10, color='skyblue', edgecolor='black')
plt.title("Students ke Grades ka Distribution")
plt.xlabel("Grade")
plt.ylabel("Kitne Students")
plt.show()
```

```python
# Graph 2 - Study Time vs Grades
plt.figure(figsize=(8,5))
sns.scatterplot(x='studytime', y='G3', data=df)
plt.title("Study Time vs Final Grade")
plt.xlabel("Study Time (1=kam, 4=zyada)")
plt.ylabel("Final Grade")
plt.show()
```

```python
# Graph 3 - Male vs Female Average Score
plt.figure(figsize=(6,4))
df.groupby('sex')['G3'].mean().plot(kind='bar', color=['pink','lightblue'])
plt.title("Male vs Female Average Score")
plt.xlabel("Gender (F=Female, M=Male)")
plt.ylabel("Average Grade")
plt.xticks(rotation=0)
plt.show()
```
